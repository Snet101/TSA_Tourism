```python
import pandas as pd

#loading merged monthly file
df = pd.read_csv("data_clean/final_monthly_311_plus_trends.csv")

df["month"] = pd.to_datetime(df["month"])#making sure month is treated as a date

corr_vars = df[["n_311_calls", "n_tourism_calls", "tourism_call_share", "tourism_intensity"]]
corr_table = corr_vars.corr().round(2)
corr_table.to_csv("data_clean/correlation_table_overall.csv")
print(corr_table)
```

```python
# correlation table PER CITY 
per_city_corr = (
    df.groupby("city")[["n_311_calls", "n_tourism_calls", "tourism_call_share", "tourism_intensity"]]
      .corr()
      .round(2)
)

per_city_corr.to_csv("data_clean/correlation_table_by_city.csv")
print(per_city_corr)
```

```python
#clean table for tsa document
per_city_corr_clean = per_city_corr.reset_index()
per_city_corr_clean.to_csv(
    "data_clean/correlation_table_by_city_clean.csv",
    index=False
)
```
